```python
import pandas as pd
import numpy as np

def remove_highly_correlated(data, threshold=0.8):
    corr_matrix = data.corr().abs()
    upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    # Identifier les colonnes à supprimer
    to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > threshold)]

    # Afficher les colonnes supprimées pour vérification
    print("Variables supprimées à cause de la corrélation forte (> {:.2f}) :\n{}".format(threshold, to_drop))

    return data.drop(columns=to_drop), to_drop

```

```python
def calculate_ranking_error(true_ranking, predicted_ranking):
    true_ranking_sorted = sorted(true_ranking, key=lambda x: x[1], reverse=True)
    predicted_ranking_sorted = sorted(predicted_ranking, key=lambda x: x[1], reverse=True)

    true_teams = [team for team, _ in true_ranking_sorted]
    predicted_teams = [team for team, _ in predicted_ranking_sorted]

    error_score = 0
    for true_team, predicted_team in zip(true_teams, predicted_teams):
        if true_team != predicted_team:
            error_score += 1

    return error_score
```

```python
import statsmodels.api as sm
import pandas as pd

import pandas as pd
import statsmodels.api as sm
from itertools import combinations


Y = Y_train

X = X_train

def best_subset_selection(X, Y):
    Y = Y.reset_index(drop=True)
    X = X.reset_index(drop=True)

    n = X.shape[1]
    all_combinations = []
    for r in range(1, n+1):
        all_combinations.extend(combinations(range(n), r))

    best_aic = float('inf')
    best_model = None
    best_variables = []

    for combo in all_combinations:
        X_subset = X.iloc[:, list(combo)]
        X_subset = sm.add_constant(X_subset)
        model = sm.OLS(Y, X_subset).fit()

        if model.aic < best_aic:
            best_aic = model.aic
            best_model = model
            best_variables = X.columns[list(combo)]

    return best_model, best_variables, best_aic

```
